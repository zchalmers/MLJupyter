```python
import win32com.client

def list_windows_devices():
    objWMIService = win32com.client.Dispatch("WbemScripting.SWbemLocator")
    objSWbemServices = objWMIService.ConnectServer(".", "root\\cimv2")
    colItems = objSWbemServices.ExecQuery("SELECT * FROM Win32_PnPEntity")
    for objItem in colItems:
        print(f"Device: {objItem.Name}")
        
list_windows_devices()
```

```python
import mss
import numpy as np
import time
from IPython.display import display, Image
from PIL import Image as PILImage
import io

def fast_screen_playback_v2(interval=0.01, duration=5):
    times = []
    with mss.mss() as sct:
        start_time = time.time()
        monitor = sct.monitors[1]  # Use primary monitor
        output_image = display("", display_id=True)
        frame_count = 0
        
        while time.time() - start_time < duration:
            frame_start_time = time.time()

            # Capture the screen directly into memory
            screenshot = sct.grab(monitor)
            frame = np.array(screenshot)  # Convert to NumPy array

            # Optional: Downscale resolution
            frame = PILImage.fromarray(frame).resize((800, 600))

            # Update the display directly without PNG conversion
            output_image.update(Image(data=frame.tobytes(), format='png'))

            # Track performance
            frame_time = time.time() - frame_start_time
            frame_count += 1
            print(f"Frame {frame_count}: {frame_time:.4f} seconds")
            times.append(frame_time)
            # Maintain consistent frame rate
            #time.sleep(max(0, interval - frame_time))
        
        total_time = time.time() - start_time
        print(f"Total Frames: {frame_count}")
        print(f"Average FPS: {frame_count / total_time:.2f}")

# Run the optimized function
fast_screen_playback_v2(interval=0.01, duration=5)
```
